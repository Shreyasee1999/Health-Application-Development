```python
# Function to categorize users based on activity levels
def categorize_user(row):
    if (
        row["VeryActiveMinutes"] < 4 and
        row["FairlyActiveMinutes"] < 6 and
        row["LightlyActiveMinutes"] < 127
    ):
        return "Low Activity"
    elif (
        4 <= row["VeryActiveMinutes"] <= 32 and
        6 <= row["FairlyActiveMinutes"] <= 19 and
        127 <= row["LightlyActiveMinutes"] <= 264
    ):
        return "Moderate Activity"
    else:
        return "High Activity"

# Applying the categorization function to the dataset
data["ActivityCategory"] = data.apply(categorize_user, axis=1)

# Checking the distribution of categories
category_counts = data["ActivityCategory"].value_counts()

```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
import numpy as np

# Selecting features and target for modeling
features = data[
    ["TotalSteps", "TotalDistance", "VeryActiveMinutes", "FairlyActiveMinutes",
     "LightlyActiveMinutes", "SedentaryMinutes"]
]
target = data["ActivityCategory"]

# Encoding the target variable (Low Activity: 0, Moderate Activity: 1, High Activity: 2)
target_encoded = target.map({"Low Activity": 0, "Moderate Activity": 1, "High Activity": 2})

# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target_encoded, test_size=0.2, random_state=42)

# Standardizing the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Using Random Forest Classifier for prediction
model = RandomForestClassifier(random_state=42)
model.fit(X_train_scaled, y_train)

# Making predictions and evaluating the model
y_pred = model.predict(X_test_scaled)
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred, target_names=["Low Activity", "Moderate Activity", "High Activity"])

accuracy, report
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import seaborn as sns

# 1. Plotting Feature Importance
feature_importance = model.feature_importances_
features = ["TotalSteps", "TotalDistance", "VeryActiveMinutes", "FairlyActiveMinutes",
            "LightlyActiveMinutes", "SedentaryMinutes"]

# Creating a bar plot for feature importance
plt.figure(figsize=(10, 6))
plt.barh(features, feature_importance)
plt.xlabel("Importance Score")
plt.title("Feature Importance in Random Forest Classifier")
plt.gca().invert_yaxis()
plt.show()

# 2. Plotting Confusion Matrix
# Generate predictions for the test set
conf_matrix = confusion_matrix(y_test, y_pred)

# Plotting the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=["Low", "Moderate", "High"],
            yticklabels=["Low", "Moderate", "High"])
plt.xlabel("Predicted Category")
plt.ylabel("Actual Category")
plt.title("Confusion Matrix of Random Forest Classifier")
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier
import time

# Initialize a dictionary to store the results
model_results = {}

# Helper function to fit the model and evaluate its performance
def evaluate_model(model, model_name):
    start_time = time.time()
    model.fit(X_train_scaled, y_train)
    y_pred = model.predict(X_test_scaled)
    end_time = time.time()

    # Calculate evaluation metrics
    accuracy = accuracy_score(y_test, y_pred)
    report = classification_report(y_test, y_pred, target_names=["Low Activity", "Moderate Activity", "High Activity"], output_dict=True)
    training_time = end_time - start_time

    # Store the results
    model_results[model_name] = {
        "Accuracy": accuracy,
        "Training Time (s)": training_time,
        "F1-Score (Low Activity)": report["Low Activity"]["f1-score"],
        "F1-Score (Moderate Activity)": report["Moderate Activity"]["f1-score"],
        "F1-Score (High Activity)": report["High Activity"]["f1-score"]
    }

# 1. Logistic Regression
logistic_model = LogisticRegression(random_state=42)
evaluate_model(logistic_model, "Logistic Regression")

# 2. Support Vector Machine (SVM)
svm_model = SVC(kernel='rbf', random_state=42)
evaluate_model(svm_model, "Support Vector Machine")

# 3. Gradient Boosting Classifier
gradient_boosting_model = GradientBoostingClassifier(random_state=42)
evaluate_model(gradient_boosting_model, "Gradient Boosting Classifier")

# 4. Comparing with the existing Random Forest Classifier
evaluate_model(model, "Random Forest Classifier")

# Converting the results to a DataFrame for easier comparison
results_df = pd.DataFrame(model_results).T

```

```python
# Code for generating and displaying confusion matrices for all models

from sklearn.metrics import confusion_matrix

# Helper function to plot confusion matrix for a given model
def plot_confusion_matrix(model, model_name):
    # Generate predictions
    y_pred = model.predict(X_test_scaled)

    # Create confusion matrix
    cm = confusion_matrix(y_test, y_pred)

    # Plot the confusion matrix
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=["Low", "Moderate", "High"],
                yticklabels=["Low", "Moderate", "High"])
    plt.title(f"Confusion Matrix for {model_name}")
    plt.xlabel("Predicted Category")
    plt.ylabel("Actual Category")
    plt.show()

# Plot confusion matrices for all models
plot_confusion_matrix(logistic_model, "Logistic Regression")
plot_confusion_matrix(svm_model, "Support Vector Machine")
plot_confusion_matrix(gradient_boosting_model, "Gradient Boosting Classifier")
plot_confusion_matrix(model, "Random Forest Classifier")
```

```python
from scipy.stats import linregress

# Function to calculate trend (slope) for TotalSteps and VeryActiveMinutes for each user
def calculate_user_trends(user_data):
    # Calculate the slope of TotalSteps and VeryActiveMinutes over time
    step_slope, _, _, _, _ = linregress(range(len(user_data)), user_data["TotalSteps"])
    active_minutes_slope, _, _, _, _ = linregress(range(len(user_data)), user_data["VeryActiveMinutes"])
    return step_slope, active_minutes_slope

# Group the data by user Id and apply the trend calculation
trends = []
for user_id, user_data in data.groupby("Id"):
    user_data_sorted = user_data.sort_values(by="ActivityDate")
    step_slope, active_minutes_slope = calculate_user_trends(user_data_sorted)
    trends.append([user_id, step_slope, active_minutes_slope])

# Create a DataFrame with the trend results
trends_df = pd.DataFrame(trends, columns=["Id", "StepSlope", "ActiveMinutesSlope"])

# Identifying plateaued users (negative slopes for both TotalSteps and VeryActiveMinutes)
plateaued_users = trends_df[(trends_df["StepSlope"] < 0) & (trends_df["ActiveMinutesSlope"] < 0)]

# Attempting to visualize the trends for plateaued users using a scatter plot

# Scatter plot for Step Slope vs. Active Minutes Slope
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x=trends_df["StepSlope"],
    y=trends_df["ActiveMinutesSlope"],
    color='blue', alpha=0.6, label="All Users"
)
sns.scatterplot(
    x=plateaued_users["StepSlope"],
    y=plateaued_users["ActiveMinutesSlope"],
    color='red', alpha=0.8, label="Plateaued Users"
)
plt.axvline(0, color='black', linestyle='--', label="Zero Slope (No Change)")
plt.axhline(0, color='black', linestyle='--')
plt.title("Trend Analysis: Step Slope vs. Active Minutes Slope")
plt.xlabel("Step Slope (Trend of Total Steps)")
plt.ylabel("Active Minutes Slope (Trend of Very Active Minutes)")
plt.legend()
plt.tight_layout()
plt.show()
```
